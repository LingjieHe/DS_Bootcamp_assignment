8- Create a DataFrame whose index is the first 10 letters of the alphabet and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.


```python
import pandas as pd
import string
import sympy

def fibonacci(num): 
    if num>=2:
        fbn = [0, 1]
        for i in range(2,num):
            fbn.append(fbn[i-2]+fbn[i-1])
        return fbn
    elif num==1:
        return [0]
 

letters = [string.ascii_letters[i] for i in range(10)]
prime = [sympy.prime(i) for i in range(1, 11)]
fibonacciNum = fibonacci(10)
dataf = pd.DataFrame(index=letters, columns=['prime','fibonacciNum'])
dataf['prime'] = prime
dataf['fibonacciNum'] = fibonacciNum

```
