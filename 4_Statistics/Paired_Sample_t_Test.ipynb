3. In your chosen state, which one of the lessons the students are more successful, math or reading?



```python
ca
```

```python
ca.describe()
```

```python
from scipy import stats

t4, p4 = stats.ttest_ind(
    ca['AVG_MATH_4_SCORE'].dropna(), ca['AVG_READING_4_SCORE'].dropna())
print("p value between math and reading at 4th grade: ", p4)

t8, p8 = stats.ttest_ind(
    ca['AVG_MATH_8_SCORE'].dropna(), ca['AVG_READING_8_SCORE'].dropna())
print("p value between math and reading at 8th grade: ", p8)

# math is more successful
```
