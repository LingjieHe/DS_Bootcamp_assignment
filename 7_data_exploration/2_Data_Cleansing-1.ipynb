```python
import pandas as pd

movie = pd.read_csv('Datasets/movies.csv')
movie.head()
```

```python
movie.describe()
```

```python
movie.info(verbose=False)
```

```python
# Year: interval variable; rating: ratio variable
def category(rating):
    if rating >= 8:
        return 1
    elif (rating <8) and (rating>=6):
        return 2
    else:
        return 3


movie['ratingNum'] = movie['Rating'].apply(category)
```

```python
movie['ratingNum'].value_counts()
```

```python
movie.groupby('ratingNum')['Rating'].count()
```

```python
# from Soudeh
variable_type = pd.Series(['continouse numerical',
                           'string variable',
                           'nominal categorical',
                           'string variable',
                           'nominal categorical',
                           'nominal categorical',
                           'discrete numerical',
                           'discrete numerical',
                           'continuous numerical',
                           'continuous numerical',
                           'continuous numerical',
                           'continuous numerical'], index=movie.dtypes.index)
movie_dtype = pd.concat([movie.dtypes, movie.nunique(), variable_type], axis=1)
movie_dtype.columns = ['data type', 'No. of categories', 'variable_type']
movie_dtype
```
