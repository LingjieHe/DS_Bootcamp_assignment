```python
import pandas as pd

edu_org = pd.read_csv('Datasets/states_all.csv')
edu_org.head()
```

```python
#From Soudeh
edu = edu_org.copy(deep=True)
edu.dtypes
edu.isnull().sum()
round(edu.isnull().sum()*100/len(edu),2)
```

```python
edu_datatype = pd.concat([edu.dtypes, edu.nunique(), round(edu.nunique()*100/len(edu), 2),
                          edu.isnull().sum(), round(edu.isnull().sum()*100/len(edu), 2)], axis=1)
edu_datatype.columns = ['data type', 'No. of categories',
                        'No. of categories %', 'missing values', 'missing values %']
edu_datatype.sort_values(by='missing values %', ascending=False, inplace=True)
```

```python
edu_datatype = edu_datatype.loc[edu_datatype['missing values']!=0]
edu_datatype
```

```python
import numpy as np

edu_datatype.iloc[np.where(edu_datatype['missing values %'] > 50)]
```

```python
columns_to_drop = edu_datatype.iloc[np.where(edu_datatype['missing values %'] > 50)].index
columns_to_drop.to_list()
```

```python
edu.drop(columns_to_drop.to_list(), axis=1, inplace=True)
edu.head(2)
```

```python
edu_datatype.drop(index=columns_to_drop, axis=0, inplace=True)
edu.describe()
```

```python
edu_impute_median = edu.copy(deep=True)
for col in edu.columns:
    if col in edu_datatype.index:
        edu_impute_median[col] = edu_impute_median[col].fillna(edu_impute_median[col].median())
edu_impute_median.isnull().sum()
```

```python
# my original method didn't drop the AVG columns and use mean to fill NaN, but here using the dataset after dropping NaN
edu_years = edu.groupby('STATE').mean()
for col in edu_years.columns:
    edu_years[col] = edu_years[col].fillna(edu_years[col].mean())
edu_years.isnull().sum()
```

4, This time, fill in the missing values using interpolation (extrapolation).  


```python
edu_interpolate = edu_org.copy(deep=True)
for col in edu.columns:
    if col in edu_datatype.index:
        edu_interpolate[col]=edu_interpolate[col].interpolate(method='linear', limit_direction='forward')
```
